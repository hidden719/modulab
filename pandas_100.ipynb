# Apply, Map (56~)

```python
df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv',index_col=0)
display(df.head(3))
df.shape
```

```python
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'   
}

df['newIncome']  =df.Income_Category.map(lambda x: dic[x])

df['newIncome']
```

```python
def changeCategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e'

df['newIncome']  =df.Income_Category.apply(changeCategory)

df['newIncome']
```

```python
df['AgeState']  = df.Customer_Age.map(lambda x: x//10 *10)

df['AgeState'].value_counts().sort_index()
```

```python
df['newEduLevel'] = df.Education_Level.map(lambda x : 1 if 'Graduate' in x else 0)
df['newEduLevel'].value_counts()
```

```python
df['newLimit'] = df.Credit_Limit.map(lambda x : 1 if x>=4500 else 0)
df['newLimit'].value_counts()
```

```python
def check(x):
    if x.Marital_Status =='Married' and x.Card_Category =='Platinum':
        return 1
    else:
        return 0


df['newState'] = df.apply(check,axis=1)

df['newState'].value_counts()
```

```python
def changeGender(x):
    if x =='M':
        return 'male'
    else:
        return 'female'
df['Gender'] = df.Gender.apply(changeGender)
df['Gender'].value_counts()
```

## Types of fillna methods
Scalar Value:

fillna(value): Fills missing values with a specified scalar value.
python
Copy code
df.fillna(0)
Forward Fill (ffill) and Backward Fill (bfill):

fillna(method='ffill'): Fills missing values with the previous non-null value (forward fill).
fillna(method='bfill'): Fills missing values with the next non-null value (backward fill).
python
Copy code
df.fillna(method='ffill')
df.fillna(method='bfill')
Fill with Mean, Median, or Mode:

fillna(value=df['column'].mean()): Fills missing values with the mean of the column.
fillna(value=df['column'].median()): Fills missing values with the median of the column.
fillna(value=df['column'].mode()[0]): Fills missing values with the mode of the column.
python
Copy code
df['column'].fillna(value=df['column'].mean())
Fill with Interpolation:

fillna(method='interpolate'): Fills missing values using linear interpolation.
python
Copy code
df.interpolate()
Limit the Number of Consecutive NaN Values Filled:

fillna(method='ffill', limit=n): Limits the number of consecutive missing values filled in a forward fill to n.
fillna(method='bfill', limit=n): Limits the number of consecutive missing values filled in a backward fill to n.
python
Copy code
df.fillna(method='ffill', limit=2)
These are just a few examples of the types of fillna methods available in pandas. The method you choose depends on the specific requirements of your data and analysis.

```python
df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')
```

```python
df.head()
```

```python
def change_date(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(":")[0]
    
    if hour =='24':
        hour ='00:00:00'
        
        FinalDate = pd.to_datetime(date +" "+hour) +datetime.timedelta(days=1) #days 만큼의 날들 후의 날짜
        
    else:
        hour = hour +':00:00'
        FinalDate = pd.to_datetime(date +" "+hour)
    
    return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)

Ans = df
Ans
```

```python
df['dayName']  =df['(년-월-일:시)'].dt.day_name()
Ans =df['dayName']
Ans.head()
```

```python
Ans1 = df.groupby(['dayName','PM10등급'],as_index=False).size()
Ans2 = Ans1.pivot(index='dayName',columns='PM10등급',values='size').fillna(0)
Ans2
```
